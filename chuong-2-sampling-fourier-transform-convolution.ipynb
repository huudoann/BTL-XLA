```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Hàm tính SNR (Signal-to-Noise Ratio)
def calculate_snr(image_array):
    mean = np.mean(image_array)
    std = np.std(image_array)
    return mean / std if std != 0 else 0

# Đọc ảnh từ đường dẫn đã chỉ định
img = Image.open("/kaggle/input/xla-images/elephant_g.jpg")
img1 = Image.open("/kaggle/input/xla-images/aero.jpg")
img2 = Image.open("/kaggle/input/xla-images/earthfromsky.jpg")

# Tạo danh sách chứa các bức ảnh
list_img = [img, img1, img2]

# Các số màu khác nhau
num_colors_list = [1 << n for n in range(8, 0, -1)]

# Điều chỉnh số lượng subplot dựa trên tổng số ảnh sẽ hiển thị
total_images = len(num_colors_list) * len(list_img)
plt.figure(figsize=(20, 40))  # Tăng chiều cao để vừa với nhiều hàng hơn
snr_list = []
i = 1

# Lặp qua các số màu
for num_colors in num_colors_list:
    for img in list_img:
        # Chuyển đổi ảnh thành ảnh với palette màu thích nghi với số màu nhất định
        im1 = img.convert('P', palette=Image.ADAPTIVE, colors=num_colors)
        
        # Chuyển ảnh sang mảng numpy để tính SNR
        im_array = np.array(im1)
        snr = calculate_snr(im_array)
        snr_list.append(snr)

        # Hiển thị ảnh và thêm tiêu đề
        plt.subplot(total_images // 2, 2, i)  # Tăng số hàng lên để chứa tất cả ảnh
        plt.imshow(im1)
        plt.axis('off')
        plt.title(f'Image with # colors = {num_colors} SNR = {np.round(snr, 3)}', size=15)
        
        i += 1

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.2, hspace=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Mở bức ảnh từ đường dẫn đã chỉ định

img = Image.open("/kaggle/input/xla-images/elephant_g.jpg").convert('L')
img1 = Image.open("/kaggle/input/xla-images/aero.jpg").convert('L')
img2 = Image.open("/kaggle/input/xla-images/earthfromsky.jpg").convert('L')

# Tạo danh sách chứa các bức ảnh
list_img = [img, img1, img2]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Vòng lặp để hiển thị từng bức ảnh và phổ tần số của chúng
for i in range(3):
    # Lấy bức ảnh tại chỉ số i trong danh sách và chuyển đổi thành mảng numpy
    img_array = np.array(list_img[i])
    
    # Tính DFT của ảnh và dịch chuyển thành phần tần số thấp vào trung tâm
    dft = np.fft.fft2(img_array)
    dft_shift = np.fft.fftshift(dft)
    
    # Lấy giá trị phổ tần số (magnitude spectrum)
    magnitude_spectrum = 20 * np.log(np.abs(dft_shift) + 1)  # Thêm 1 để tránh log(0)
    
    # Tạo một subplot cho ảnh gốc
    plt.subplot(2, 3, i + 1)
    plt.imshow(img_array, cmap='gray')
    plt.title(f'Original Image {i + 1}', fontsize=20)
    plt.axis('off')
    
    # Tạo một subplot cho phổ tần số
    plt.subplot(2, 3, i + 4)
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title(f'Magnitude Spectrum {i + 1}', fontsize=20)
    plt.axis('off')
    
# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.1, hspace=0.3)

# Hiển thị tất cả các bức ảnh và phổ tần số trong cửa sổ
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy import signal
import scipy.misc as misc

# Mở bức ảnh từ đường dẫn đã chỉ định và chuẩn hóa giá trị pixel vào khoảng [0, 1]
img = Image.open("/kaggle/input/xla-images/earthfromsky.jpg")
img = np.array(img) / 255.0

print(np.max(img))  # Kiểm tra giá trị lớn nhất sau khi chuẩn hóa
print(img.shape)    # Kiểm tra kích thước của ảnh

# Khởi tạo kernel cho hiệu ứng emboss và phát hiện cạnh Scharr
emboss_kernel = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])
edge_schar_kernel = np.array([[-3-3j, 0-10j, 3-3j],
                              [-10+0j, 0+0j, 10+0j],
                              [-3+3j, 0+10j, 3+3j]])

# Khởi tạo mảng cho ảnh đã qua xử lý emboss và phát hiện cạnh
im_embossed = np.ones(img.shape)
im_edges = np.ones(img.shape)

# Áp dụng bộ lọc emboss cho từng kênh màu
for i in range(3):
    im_embossed[..., i] = np.clip(signal.convolve2d(img[..., i], emboss_kernel, mode='same', boundary='symm'), 0, 1)

# Áp dụng bộ lọc Scharr cho từng kênh màu
for i in range(3):
    im_edges[..., i] = np.clip(np.real(signal.convolve2d(img[..., i], edge_schar_kernel, mode='same', boundary='symm')), 0, 1)

# Hiển thị ảnh gốc, ảnh emboss, và ảnh phát hiện cạnh
fig, axes = plt.subplots(nrows=3, figsize=(20, 30))

# Ảnh gốc
axes[0].imshow(img)
axes[0].set_title('Original Image', size=20)

# Ảnh đã qua xử lý emboss
axes[1].imshow(im_embossed)
axes[1].set_title('Embossed Image', size=20)

# Ảnh phát hiện cạnh Scharr
axes[2].imshow(im_edges)
axes[2].set_title('Scharr Edge Detection', size=20)

# Tắt trục tọa độ để hình ảnh hiển thị đẹp hơn
for ax in axes:
    ax.axis('off')

# Hiển thị tất cả các bức ảnh
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy import signal

# Mở các bức ảnh từ đường dẫn đã chỉ định và chuyển sang ảnh xám (grayscale)
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Lặp qua từng ảnh để xử lý
for img_path in img_paths:
    # Đọc ảnh và chuyển sang grayscale
    img = Image.open(img_path).convert('L')
    img = np.array(img) - np.mean(np.array(img))  # Chuẩn hóa ảnh

    # Lấy một mẫu ảnh nhỏ từ ảnh chính (giả sử phần từ [100:150, 100:150])
    template_image = np.copy(img[100:150, 100:150])
    template_image -= np.mean(template_image)

    # Thêm nhiễu ngẫu nhiên vào ảnh gốc
    noisy_image = img + np.random.randn(*img.shape) * 50

    # Tính tương quan chéo 2D giữa ảnh gốc và ảnh mẫu
    correlation = signal.correlate2d(noisy_image, template_image, boundary='symm', mode='same')

    # Tìm tọa độ của điểm có giá trị cao nhất trong kết quả tương quan
    y, x = np.unravel_index(np.argmax(correlation), correlation.shape)

    # Hiển thị ảnh gốc, ảnh mẫu, và kết quả tương quan
    fig, (ax_original, ax_template, ax_correlation) = plt.subplots(3, 1, figsize=(6, 15))

    # Hiển thị ảnh gốc (có nhiễu)
    ax_original.imshow(noisy_image, cmap='gray')
    ax_original.set_title(f'Original (with noise) - {img_path}', size=20)
    ax_original.set_axis_off()
    ax_original.plot(x, y, 'ro')  # Đánh dấu vị trí tìm thấy bằng chấm đỏ

    # Hiển thị ảnh mẫu
    ax_template.imshow(template_image, cmap='gray')
    ax_template.set_title('Template', size=20)
    ax_template.set_axis_off()

    # Hiển thị ảnh tương quan chéo
    ax_correlation.imshow(correlation, cmap='afmhot')
    ax_correlation.set_title('Cross-correlation', size=20)
    ax_correlation.set_axis_off()

    # Hiển thị các hình ảnh
    plt.tight_layout()
    plt.show()
```
