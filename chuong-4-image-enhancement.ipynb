```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Mở các bức ảnh từ đường dẫn
img = Image.open("/kaggle/input/xla-images/elephant_g.jpg")
img1 = Image.open("/kaggle/input/xla-images/aero.jpg")
img2 = Image.open("/kaggle/input/xla-images/earthfromsky.jpg")

# Tạo danh sách chứa các bức ảnh
list_img = [img, img1, img2]

# Thiết lập tham số gamma cho phép biến đổi lũy thừa
gamma = 2.5

# Hàm thực hiện phép biến đổi lũy thừa
def power_law_transform(image, gamma):
    # Chuyển ảnh sang numpy array và chuẩn hóa
    img_array = np.array(image) / 255.0
    # Áp dụng phép biến đổi lũy thừa
    transformed_img = np.power(img_array, gamma)
    # Chuyển lại về khoảng [0, 255] và định dạng kiểu uint8
    return (transformed_img * 255).astype(np.uint8)

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Vòng lặp để hiển thị từng bức ảnh sau khi biến đổi lũy thừa
for i in range(3):
    # Tạo một subplot cho mỗi bức ảnh
    plt.subplot(1, 3, i + 1)
    
    # Áp dụng phép biến đổi lũy thừa
    transformed_img = power_law_transform(list_img[i], gamma)
    
    # Hiển thị bức ảnh sau khi biến đổi
    plt.imshow(transformed_img)
    
    # Tắt hiển thị trục tọa độ x và y để ảnh nhìn đẹp hơn
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# Hiển thị tất cả các bức ảnh trong cửa sổ
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Mở các bức ảnh từ đường dẫn
img = Image.open("/kaggle/input/xla-images/elephant_g.jpg")
img1 = Image.open("/kaggle/input/xla-images/aero.jpg")
img2 = Image.open("/kaggle/input/xla-images/earthfromsky.jpg")

# Tạo danh sách chứa các bức ảnh
list_img = [img, img1, img2]

# Hàm điều chỉnh độ tương phản tuyến tính từng đoạn
def contrast(c):
    return 0 if c < 70 else (255 if c > 150 else (255 * c - 22950) / 48)

# Hàm áp dụng điều chỉnh độ tương phản
def apply_contrast(image):
    # Áp dụng hàm contrast cho từng điểm ảnh
    return image.point(contrast)

# Hàm hiển thị biểu đồ histogram
def plot_hist(r, g, b, title):
    plt.hist(r.ravel(), bins=256, color='red', alpha=0.5, label='Đỏ')
    plt.hist(g.ravel(), bins=256, color='green', alpha=0.5, label='Xanh lá')
    plt.hist(b.ravel(), bins=256, color='blue', alpha=0.5, label='Xanh dương')
    plt.legend(loc='upper right')
    plt.title(title)
    plt.xlabel('Cường độ điểm ảnh')
    plt.ylabel('Tần suất')
    plt.yscale('log')

# Sử dụng phong cách đồ thị ggplot
plt.style.use('ggplot')

# Tạo cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(15, 5))

for i in range(3):
    # Áp dụng điều chỉnh độ tương phản
    im1 = apply_contrast(list_img[i])

    # Tách các kênh màu
    im_r, im_g, im_b = im1.split()[:3]

    # Hiển thị ảnh sau khi điều chỉnh độ tương phản
    plt.subplot(2, 3, i + 1)
    plt.imshow(im1)
    plt.axis('off')

    # Hiển thị biểu đồ histogram của ảnh đã điều chỉnh
    plt.subplot(2, 3, i + 4)
    plot_hist(np.array(im_r), np.array(im_g), np.array(im_b), f'Histogram của ảnh {i+1} sau khi điều chỉnh')

# Hiển thị tất cả các hình ảnh và histogram
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh và chuyển đổi thành ảnh grayscale
img = Image.open(img_path).convert('L')  # Chuyển thành ảnh xám
img_array = np.array(img, dtype=np.float32)  # Chuyển thành mảng để dễ xử lý

# Lấy kích thước của ảnh
height, width = img_array.shape

# Floyd-Steinberg dithering với sự khuếch tán lỗi
for y in range(height):
    for x in range(width):
        old_pixel = img_array[y, x]
        new_pixel = 255 if old_pixel > 127 else 0  # Chuyển đổi thành trắng hoặc đen
        img_array[y, x] = new_pixel
        quant_error = old_pixel - new_pixel  # Tính lỗi

        # Phân phối lỗi đến các pixel lân cận
        if x + 1 < width:
            img_array[y, x + 1] += quant_error * 7 / 16
        if x - 1 >= 0 and y + 1 < height:
            img_array[y + 1, x - 1] += quant_error * 3 / 16
        if y + 1 < height:
            img_array[y + 1, x] += quant_error * 5 / 16
        if x + 1 < width and y + 1 < height:
            img_array[y + 1, x + 1] += quant_error * 1 / 16

# Chuyển mảng kết quả thành ảnh và hiển thị
dithered_img = Image.fromarray(img_array.astype(np.uint8))

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(12, 8))

# Hiển thị ảnh đã dithered
plt.imshow(dithered_img, cmap='gray')  # Đặt `cmap='gray'` để hiển thị ảnh đen trắng
plt.axis('off')  # Tắt hiển thị trục tọa độ x và y để ảnh nhìn đẹp hơn

# Hiển thị ảnh
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure, io, img_as_float
import matplotlib

# Cấu hình font size của matplotlib
matplotlib.rcParams['font.size'] = 8

# Định nghĩa hàm để hiển thị ảnh và histogram với CDF
def plot_image_and_hist(image, axes, bins=256):
    image = img_as_float(image)  # Chuyển đổi ảnh thành float để xử lý
    axes_image, axes_hist = axes
    axes_cdf = axes_hist.twinx()
    
    # Hiển thị ảnh
    axes_image.imshow(image, cmap='gray')
    axes_image.set_axis_off()
    
    # Hiển thị histogram
    axes_hist.hist(image.ravel(), bins=bins, histtype='step', color='black')
    axes_hist.set_xlim(0, 1)
    axes_hist.set_xlabel('Pixel intensity', size=15)
    axes_hist.ticklabel_format(axis='y', style='scientific', scilimits=(0, 0))
    axes_hist.set_yticks([])
    
    # Hiển thị CDF (Cumulative Distribution Function)
    image_cdf, bins = exposure.cumulative_distribution(image, bins)
    axes_cdf.plot(bins, image_cdf, 'r')
    axes_cdf.set_yticks([])
    
    return axes_image, axes_hist, axes_cdf

# Đọc ảnh đầu vào
img_path = "/kaggle/input/xla-images/elephant_g.jpg"
im = io.imread(img_path)

# Áp dụng các kỹ thuật cải thiện ảnh
im_rescale = exposure.rescale_intensity(im, in_range=(im.min(), im.max()), out_range=(0, 1))
im_eq = exposure.equalize_hist(im)  # Histogram equalization
im_adapteq = exposure.equalize_adapthist(im, clip_limit=0.03)  # Adaptive histogram equalization

# Tạo các subplots để hiển thị các kết quả khác nhau
fig = plt.figure(figsize=(15, 7))
axes = np.zeros((2, 4), dtype=object)
axes[0, 0] = fig.add_subplot(2, 4, 1)

# Tạo các subplot cho từng kỹ thuật
for i in range(1, 4):
    axes[0, i] = fig.add_subplot(2, 4, 1 + i, sharex=axes[0, 0], sharey=axes[0, 0])
for i in range(4):
    axes[1, i] = fig.add_subplot(2, 4, 5 + i)

# Hiển thị từng ảnh và histogram tương ứng
axes_image, axes_hist, axes_cdf = plot_image_and_hist(im, axes[:, 0])
axes_image.set_title('Low contrast image', size=20)
y_min, y_max = axes_hist.get_ylim()
axes_hist.set_ylabel('Number of pixels', size=20)
axes_hist.set_yticks(np.linspace(0, y_max, 5))

axes_image, axes_hist, axes_cdf = plot_image_and_hist(im_rescale, axes[:, 1])
axes_image.set_title('Contrast stretching', size=20)

axes_image, axes_hist, axes_cdf = plot_image_and_hist(im_eq, axes[:, 2])
axes_image.set_title('Histogram equalization', size=20)

axes_image, axes_hist, axes_cdf = plot_image_and_hist(im_adapteq, axes[:, 3])
axes_image.set_title('Adaptive equalization', size=20)

# Thiết lập nhãn cho CDF
axes_cdf.set_ylabel('Fraction of total intensity', size=20)
axes_cdf.set_yticks(np.linspace(0, 1, 5))

# Sắp xếp bố cục và hiển thị
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
import pylab

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh
img = Image.open(img_path).convert("RGB")  # Chuyển đổi sang RGB nếu cần

# Thêm nhiễu salt-and-pepper vào ảnh với mức nhiễu cao hơn
def add_salt_and_pepper_noise(img, prob=0.3):  # Tăng giá trị prob lên để thêm nhiều nhiễu hơn
    output = np.array(img)
    noise = np.random.rand(*output.shape[:2])  # Tạo ma trận nhiễu cùng kích thước
    salt = noise < prob / 2
    pepper = noise > 1 - prob / 2
    output[salt] = [255, 255, 255]  # White (salt)
    output[pepper] = [0, 0, 0]  # Black (pepper)
    return Image.fromarray(output)

# Thêm nhiễu vào ảnh
img_noisy = add_salt_and_pepper_noise(img, prob=0.3)

# Hiển thị ảnh gốc và ảnh có nhiễu
pylab.figure(figsize=(20, 7))
pylab.subplot(1, 3, 1)
pylab.imshow(img_noisy)
pylab.title('Noisy Image (High Noise)', size=30)
pylab.axis('off')

# Hàm để hiển thị ảnh sau khi làm mờ
def plot_image(img, title):
    plt.imshow(img)
    plt.title(title, size=20)
    plt.axis('off')

# Áp dụng bộ lọc mean (box blur) với các kích thước kernel khác nhau
for i, n in enumerate([3, 5], start=2):
    box_blur_kernel = np.ones((n, n)) / (n * n)  # Kernel trung bình
    im_blurred = img_noisy.filter(ImageFilter.Kernel((n, n), box_blur_kernel.flatten()))
    pylab.subplot(1, 3, i)
    plot_image(im_blurred, f'Blurred with kernel size = {n}x{n}')

# Đặt tiêu đề cho toàn bộ
pylab.suptitle('PIL Mean Filter (Box Blur) with Different Kernel Sizes', size=30)
pylab.show()
```

```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh
img = Image.open(img_path)

# Chuyển đổi ảnh thành mảng numpy
img_array = np.array(img)

# Thêm nhiễu vào ảnh
def add_noise(image, noise_factor=0.2):
    noise = np.random.normal(loc=0.0, scale=noise_factor, size=image.shape)
    noisy_image = np.clip(image + noise * 255, 0, 255).astype(np.uint8)
    return noisy_image

# Thêm nhiễu vào ảnh
noisy_img_array = add_noise(img_array)
noisy_img = Image.fromarray(noisy_img_array)

# Hiển thị ảnh gốc và ảnh bị nhiễu
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(noisy_img)
plt.axis('off')
plt.title('Noisy Image')

plt.show()

# Áp dụng Gaussian blur với các bán kính khác nhau
plt.figure(figsize=(20, 6))
i = 1
for radius in range(1, 4):
    im1 = noisy_img.filter(ImageFilter.GaussianBlur(radius))
    plt.subplot(1, 3, i)
    plt.imshow(im1)
    plt.axis('off')
    plt.title(f'Gaussian Blur Radius = {radius}')
    i += 1

plt.suptitle('PIL Gaussian Blur with Different Radius', size=20)
plt.show()
```

```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh
img = Image.open(img_path)

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(12, 8))

# Hiển thị bức ảnh gốc
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()

# Hàm thêm nhiễu muối và hạt tiêu
def add_salt_and_pepper_noise(image, prop_noise):
    noisy_image = image.copy()
    n = int(noisy_image.width * noisy_image.height * prop_noise)
    x_coords = np.random.randint(0, noisy_image.width, n)
    y_coords = np.random.randint(0, noisy_image.height, n)

    for (x, y) in zip(x_coords, y_coords):
        noisy_image.putpixel((x, y), (0, 0, 0) if np.random.rand() < 0.5 else (255, 255, 255))
    
    return noisy_image

# Tạo cửa sổ hiển thị cho ảnh với nhiễu
plt.figure(figsize=(25, 35))
i = 1

# Thêm nhiễu với các tỷ lệ khác nhau và hiển thị ảnh
for prop_noise in np.linspace(0.05, 0.3, 3):
    noisy_img = add_salt_and_pepper_noise(img.convert('RGB'), prop_noise)
    
    # Hiển thị ảnh nhiễu
    plt.subplot(6, 4, i)
    plt.imshow(noisy_img)
    plt.axis('off')
    plt.title(f'Image with {int(100 * prop_noise)}% Noise')
    i += 1
    
    # Áp dụng bộ lọc trung vị với các kích thước khác nhau
    for sz in [3, 7, 11]:
        filtered_img = noisy_img.filter(ImageFilter.MedianFilter(size=sz))
        plt.subplot(6, 4, i)
        plt.imshow(filtered_img)
        plt.axis('off')
        plt.title(f'Median Filter Size = {sz}')
        i += 1

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage.restoration import denoise_bilateral

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh và resize nó
img = Image.open(img_path).resize((200, 150))  # Resize to a smaller size

# Hàm thêm nhiễu muối và hạt tiêu
def add_salt_and_pepper_noise(image, prop_noise):
    noisy_image = np.array(image).copy()
    n = int(noisy_image.size * prop_noise)
    x_coords = np.random.randint(0, noisy_image.shape[0], n)
    y_coords = np.random.randint(0, noisy_image.shape[1], n)

    for (x, y) in zip(x_coords, y_coords):
        noisy_image[x, y] = (0, 0, 0) if np.random.rand() < 0.5 else (255, 255, 255)

    return noisy_image

# Thêm nhiễu vào ảnh
prop_noise = 0.1
noisy = add_salt_and_pepper_noise(img.convert('RGB'), prop_noise)

# Tạo cửa sổ hiển thị cho ảnh nhiễu
plt.figure(figsize=(6, 4))
plt.imshow(noisy)
plt.axis('off')
plt.title(f'Noisy Image with {int(100 * prop_noise)}% Added Noise')
plt.show()

# Tạo cửa sổ hiển thị cho các ảnh đã lọc
plt.figure(figsize=(15, 10))
i = 1

# Thử nghiệm với các giá trị sigma cho bộ lọc bilateral
for sigma_sp in [5, 10, 20]:
    for sigma_col in [0.1, 0.25, 5]:
        plt.subplot(3, 3, i)
        denoised = denoise_bilateral(noisy, sigma_color=sigma_col, sigma_spatial=sigma_sp, channel_axis=-1)

        # Resize the denoised image for better display
        plt.imshow(denoised)
        plt.title(r'$\sigma_r=$' + str(sigma_col) + r', $\sigma_s=$' + str(sigma_sp), size=20)
        plt.axis('off')
        i += 1

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma
from skimage.metrics import peak_signal_noise_ratio as compare_psnr

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh và chuyển đổi sang float, sau đó resize
img = img_as_float(Image.open(img_path).resize((100, 75)))  # Resize thành 100x75 pixel

# Thêm nhiễu vào ảnh
sigma = 0.25
noisy = img + sigma * np.random.standard_normal(img.shape)
noisy = np.clip(noisy, 0, 1)

# Ước lượng độ lệch chuẩn của nhiễu từ ảnh nhiễu
sigma_est = estimate_sigma(noisy)  # Bỏ tham số multichannel
print("Độ lệch chuẩn của nhiễu ước lượng = {}".format(sigma_est))

# Tham số cho bộ lọc non-local means
patch_kw = dict(patch_size=5, patch_distance=6)

# Thuật toán chậm
denoise = denoise_nl_means(noisy, h=1.15 * sigma_est, fast_mode=False, **patch_kw)

# Thuật toán nhanh
denoise_fast = denoise_nl_means(noisy, h=0.8 * sigma_est, fast_mode=True, **patch_kw)

# Hàm để vẽ ảnh
def plot_image_axes(image, axes, title):
    axes.imshow(image)
    axes.axis('off')
    axes.set_title(title, size=20)

# Tạo một subplot cho các ảnh
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12), sharex=True, sharey=True)
plot_image_axes(noisy, axes[0, 0], 'Ảnh Nhiễu')
plot_image_axes(denoise, axes[0, 1], 'Non-Local Means (Chậm)')
plot_image_axes(img, axes[1, 0], 'Ảnh Gốc (Không Nhiễu)')
plot_image_axes(denoise_fast, axes[1, 1], 'Non-Local Means (Nhanh)')

# Điều chỉnh bố cục
fig.tight_layout()
plt.show()

# Giá trị chỉ số PSNR
psnr_noisy = compare_psnr(img, noisy)
psnr = compare_psnr(img, denoise)
psnr_fast = compare_psnr(img, denoise_fast)

print("PSNR (Nhiễu) = {:0.2f}".format(psnr_noisy))
print("PSNR (Chậm) = {:0.2f}".format(psnr))
print("PSNR (Nhanh) = {:0.2f}".format(psnr_fast))
```

Ảnh sau khi xử lý

```python
import numpy as np
from PIL import Image
from scipy import ndimage
import matplotlib.pyplot as plt

# Đường dẫn đến bức ảnh đầu tiên
img_path = "/kaggle/input/xla-images/elephant_g.jpg"

# Mở ảnh, chuyển sang thang độ xám và thay đổi kích thước để giảm kích thước
img = Image.open(img_path).convert('L')  # Chuyển sang thang độ xám
img = img.resize((200, 200))  # Giảm kích thước để chạy nhanh hơn
lena = np.array(img)

# Thêm nhiễu salt-and-pepper vào ảnh
noise = np.random.random(lena.shape)
lena[noise > 0.9] = 255  # Giá trị nhiễu trắng
lena[noise < 0.1] = 0    # Giá trị nhiễu đen

# Hàm để hiển thị ảnh
def plot_image(data, title):
    plt.imshow(data, cmap='gray')
    plt.title(title)
    plt.axis('off')

# Hiển thị ảnh nhiễu
plt.figure(figsize=(10, 8))
plot_image(lena, 'Noisy Image')
plt.show()

# Thiết lập cửa sổ hiển thị cho các bộ lọc phần trăm
fig = plt.figure(figsize=(12, 10))
i = 1

# Thử nghiệm với các bộ lọc phần trăm và kích thước kernel khác nhau
for p in range(25, 100, 25):
    for k in range(5, 25, 5):
        plt.subplot(3, 4, i)
        filtered = ndimage.percentile_filter(lena, percentile=p, size=(k, k))
        plot_image(filtered, f'{p} percentile, {k}x{k} kernel')
        i += 1

plt.tight_layout()
plt.show()
```
