```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy import ndimage

# Đường dẫn đến các ảnh
img_paths = ["/kaggle/input/xla-images/elephant_g.jpg",
             "/kaggle/input/xla-images/aero.jpg",
             "/kaggle/input/xla-images/earthfromsky.jpg"]

# Mở ảnh và chuyển thành thang độ xám
list_img = [Image.open(path).convert('L') for path in img_paths]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(20, 15))

# Toán tử Sobel cho đạo hàm theo trục x và y
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

# Vòng lặp để tính toán và hiển thị độ lớn và hướng của gradient cho mỗi ảnh
for i, img in enumerate(list_img):
    img_gray = np.array(img)  # Chuyển ảnh thành mảng NumPy

    # Tính đạo hàm theo trục x và y
    img_dx = ndimage.convolve(img_gray, sobel_x)
    img_dy = ndimage.convolve(img_gray, sobel_y)

    # Tính độ lớn của gradient
    img_magnitude = np.sqrt(img_dx**2 + img_dy**2)
    
    # Tính hướng của gradient (theo radian)
    img_direction = np.arctan2(img_dy, img_dx)

    # Hiển thị độ lớn và hướng của gradient
    plt.subplot(2, 3, i + 1)
    plt.imshow(img_magnitude, cmap='gray')
    plt.title('Gradient Magnitude')
    plt.axis('off')
    
    plt.subplot(2, 3, i + 4)
    plt.imshow(img_direction, cmap='hsv')  # Sử dụng colormap 'hsv' để biểu diễn hướng
    plt.title('Gradient Direction')
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.1, hspace=0.1)

# Hiển thị tất cả các bức ảnh
plt.show()
```

# Unsharp masking

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import img_as_float
from scipy import ndimage

# Định nghĩa hàm chuyển đổi từ RGB sang thang độ xám
def rgb2gray(im):
    '''
    The input image is an RGB image
    with pixel values for each channel in [0, 1]
    '''
    return np.clip(0.2989 * im[..., 0] + 0.5870 * im[..., 1] + 0.1140 * im[..., 2], 0, 1)

# Mở các bức ảnh từ đường dẫn và chuyển sang dạng float
img = img_as_float(np.array(Image.open("/kaggle/input/xla-images/elephant_g.jpg")))
img1 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/aero.jpg")))
img2 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/earthfromsky.jpg")))

# Chuyển đổi các ảnh sang thang độ xám
im_list = [rgb2gray(img), rgb2gray(img1), rgb2gray(img2)]

# Làm mờ các ảnh bằng Gaussian filter
im_blurred_list = [ndimage.gaussian_filter(im, sigma=5) for im in im_list]

# Tạo ảnh chi tiết (bằng cách trừ ảnh gốc với ảnh làm mờ)
im_detail_list = [np.clip(im - im_blurred, 0, 1) for im, im_blurred in zip(im_list, im_blurred_list)]

# Các giá trị alpha để làm sắc nét
alpha = [1, 5, 10]

# Tạo một cửa sổ hiển thị với kích thước lớn
fig, axes = plt.subplots(nrows=4, ncols=3, sharex=True, sharey=True, figsize=(20, 25))
axes = axes.ravel()

# Vòng lặp để hiển thị từng ảnh gốc, ảnh làm mờ, ảnh chi tiết, và ảnh làm sắc nét
for i in range(3):
    # Hiển thị ảnh gốc
    axes[i].imshow(im_list[i], cmap='gray')
    axes[i].set_title('Original Image ' + str(i + 1), size=15)
    axes[i].axis('off')

    # Hiển thị ảnh làm mờ
    axes[i + 3].imshow(im_blurred_list[i], cmap='gray')
    axes[i + 3].set_title('Blurred Image ' + str(i + 1) + ', sigma=5', size=15)
    axes[i + 3].axis('off')

    # Hiển thị ảnh chi tiết
    axes[i + 6].imshow(im_detail_list[i], cmap='gray')
    axes[i + 6].set_title('Detail Image ' + str(i + 1), size=15)
    axes[i + 6].axis('off')

    # Hiển thị ảnh làm sắc nét với alpha=5
    im_sharp = np.clip(im_list[i] + alpha[1] * im_detail_list[i], 0, 1)
    axes[i + 9].imshow(im_sharp, cmap='gray')
    axes[i + 9].set_title('Sharpened Image ' + str(i + 1) + ', alpha=5', size=15)
    axes[i + 9].axis('off')

# Điều chỉnh bố cục hiển thị
fig.tight_layout()

# Hiển thị tất cả các bức ảnh trong cửa sổ
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import img_as_float
from skimage.color import rgb2gray
from scipy.ndimage import sobel

# Mở các bức ảnh từ đường dẫn đã chỉ định và chuyển đổi sang float
img = img_as_float(np.array(Image.open("/kaggle/input/xla-images/elephant_g.jpg")))
img1 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/aero.jpg")))
img2 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/earthfromsky.jpg")))

# Chuyển đổi các ảnh sang thang độ xám
im_list = [rgb2gray(img), rgb2gray(img1), rgb2gray(img2)]

# Tính toán độ lớn gradient cho từng ảnh
gradient_magnitude_list = []
for im in im_list:
    # Tính toán đạo hàm bậc nhất theo trục x và y
    dx = sobel(im, axis=0)  # Đạo hàm theo x
    dy = sobel(im, axis=1)  # Đạo hàm theo y
    
    # Tính toán độ lớn của gradient
    gradient_magnitude = np.sqrt(dx**2 + dy**2)
    gradient_magnitude_list.append(gradient_magnitude)

# Tạo cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Hiển thị ảnh gradient cho từng ảnh trong danh sách
for i in range(3):
    # Tạo một subplot cho mỗi ảnh gradient
    plt.subplot(1, 3, i + 1)
    
    # Hiển thị độ lớn của gradient tại chỉ số i
    plt.imshow(gradient_magnitude_list[i], cmap='gray')
    plt.title('Gradient Magnitude Image ' + str(i + 1))
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# Hiển thị tất cả các bức ảnh gradient trong cửa sổ
plt.show()
```

# Sobel edge detector

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import img_as_float
from skimage.color import rgb2gray
from scipy.ndimage import sobel

# Mở các bức ảnh từ đường dẫn đã chỉ định và chuyển đổi sang float
img = img_as_float(np.array(Image.open("/kaggle/input/xla-images/elephant_g.jpg")))
img1 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/aero.jpg")))
img2 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/earthfromsky.jpg")))

# Chuyển đổi các ảnh sang thang độ xám
im_list = [rgb2gray(img), rgb2gray(img1), rgb2gray(img2)]

# Tạo cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(20, 18))

# Hiển thị kết quả cho từng ảnh trong danh sách
for idx, im in enumerate(im_list):
    # Tính toán đạo hàm bậc nhất theo trục x và y
    edges_x = sobel(im, axis=0)  # Đạo hàm theo x
    edges_y = sobel(im, axis=1)  # Đạo hàm theo y
    
    # Tính toán độ lớn của gradient
    edges = np.sqrt(edges_x**2 + edges_y**2)

    # Hiển thị ảnh gốc
    plt.subplot(3, 4, idx * 4 + 1)
    plt.imshow(im, cmap='gray')
    plt.title('Original Image ' + str(idx + 1))
    plt.axis('off')
    
    # Hiển thị gradient theo trục x
    plt.subplot(3, 4, idx * 4 + 2)
    plt.imshow(edges_x, cmap='gray')
    plt.title('Sobel X ' + str(idx + 1))
    plt.axis('off')
    
    # Hiển thị gradient theo trục y
    plt.subplot(3, 4, idx * 4 + 3)
    plt.imshow(edges_y, cmap='gray')
    plt.title('Sobel Y ' + str(idx + 1))
    plt.axis('off')
    
    # Hiển thị độ lớn của gradient
    plt.subplot(3, 4, idx * 4 + 4)
    plt.imshow(edges, cmap='gray')
    plt.title('Gradient Magnitude ' + str(idx + 1))
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.1, hspace=0.1)

# Hiển thị tất cả các kết quả trong cửa sổ
plt.show()
```

# Different edge detectors Prewitt, Roberts, Sobel, Scharr, and Laplace

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import img_as_float
from skimage.color import rgb2gray
from skimage import filters

# Mở các bức ảnh từ đường dẫn đã chỉ định và chuyển đổi sang float
img = img_as_float(np.array(Image.open("/kaggle/input/xla-images/elephant_g.jpg")))
img1 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/aero.jpg")))
img2 = img_as_float(np.array(Image.open("/kaggle/input/xla-images/earthfromsky.jpg")))

# Chuyển đổi các ảnh sang thang độ xám
im_list = [rgb2gray(img), rgb2gray(img1), rgb2gray(img2)]

# Tạo cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(20, 24))

# Vòng lặp để hiển thị từng ảnh và các bộ lọc cạnh tương ứng
for idx, im in enumerate(im_list):
    # Hiển thị ảnh gốc
    plt.subplot(3, 6, idx * 6 + 1)
    plt.imshow(im, cmap='gray')
    plt.title('Original Image ' + str(idx + 1))
    plt.axis('off')
    
    # Roberts
    edges_roberts = filters.roberts(im)
    plt.subplot(3, 6, idx * 6 + 2)
    plt.imshow(edges_roberts, cmap='gray')
    plt.title('Roberts ' + str(idx + 1))
    plt.axis('off')
    
    # Scharr
    edges_scharr = filters.scharr(im)
    plt.subplot(3, 6, idx * 6 + 3)
    plt.imshow(edges_scharr, cmap='gray')
    plt.title('Scharr ' + str(idx + 1))
    plt.axis('off')
    
    # Sobel
    edges_sobel = filters.sobel(im)
    plt.subplot(3, 6, idx * 6 + 4)
    plt.imshow(edges_sobel, cmap='gray')
    plt.title('Sobel ' + str(idx + 1))
    plt.axis('off')
    
    # Prewitt
    edges_prewitt = filters.prewitt(im)
    plt.subplot(3, 6, idx * 6 + 5)
    plt.imshow(edges_prewitt, cmap='gray')
    plt.title('Prewitt ' + str(idx + 1))
    plt.axis('off')
    
    # Laplace
    edges_laplace = np.clip(filters.laplace(im), 0, 1)
    plt.subplot(3, 6, idx * 6 + 6)
    plt.imshow(edges_laplace, cmap='gray')
    plt.title('Laplace ' + str(idx + 1))
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.1, hspace=0.03)

# Hiển thị tất cả các kết quả trong cửa sổ
plt.show()
```

# The Canny edge detector

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import feature

# Đường dẫn đến các bức ảnh
img_paths = ["/kaggle/input/xla-images/elephant_g.jpg",
             "/kaggle/input/xla-images/aero.jpg",
             "/kaggle/input/xla-images/earthfromsky.jpg"]

# Tạo một cửa sổ hiển thị lớn
plt.figure(figsize=(30, 12))

# Vòng lặp qua từng ảnh
for i, path in enumerate(img_paths):
    # Mở ảnh, chuyển sang ảnh xám và chuyển đổi thành mảng numpy
    img = Image.open(path).convert('L')
    img_np = np.array(img)

    # Áp dụng bộ lọc Gaussian và thêm nhiễu ngẫu nhiên
    img_np = ndimage.gaussian_filter(img_np, sigma=4)
    noisy_img = img_np + 0.05 * np.random.random(img_np.shape)

    # Áp dụng bộ lọc Canny với hai giá trị sigma khác nhau
    edges_sigma1 = feature.canny(noisy_img, sigma=1)
    edges_sigma3 = feature.canny(noisy_img, sigma=3)

    # Tạo các subplots
    fig, (axes1, axes2, axes3) = plt.subplots(nrows=1, ncols=3, figsize=(30, 12), sharex=True, sharey=True)

    # Hiển thị ảnh đã thêm nhiễu
    axes1.imshow(noisy_img, cmap='gray')
    axes1.axis('off')
    axes1.set_title('Noisy Image', fontsize=20)

    # Hiển thị cạnh phát hiện được với sigma=1
    axes2.imshow(edges_sigma1, cmap='gray')
    axes2.axis('off')
    axes2.set_title('Canny Filter, $\sigma=1$', fontsize=20)

    # Hiển thị cạnh phát hiện được với sigma=3
    axes3.imshow(edges_sigma3, cmap='gray')
    axes3.axis('off')
    axes3.set_title('Canny Filter, $\sigma=3$', fontsize=20)

    plt.show()
```

Ảnh sau khi xử lý

```python
import numpy as np
from scipy.signal import convolve2d
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.pylab as pylab

# Hàm hiển thị kernel
def plot_kernel(kernel, title):
    pylab.imshow(kernel, cmap='YlOrRd')
    ax = pylab.gca()
    ax.set_xticks(np.arange(-0.5, kernel.shape[0], 2.5))
    ax.set_yticks(np.arange(-0.5, kernel.shape[1], 2.5))
    pylab.colorbar()
    pylab.title(title, size=20)

# Hàm tạo kernel LoG
def LOG(k=12, s=3):
    n = 2 * k + 1
    kernel = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            kernel[i, j] = -(1 - ((i - k)**2 + (j - k)**2) / (2. * s**2)) * \
                np.exp(-((i - k)**2 + (j - k)**2) / (2. * s**2)) / (np.pi * s**4)
    kernel = np.round(kernel / np.sqrt((kernel**2).sum()), 3)
    return kernel

# Hàm tạo kernel DoG
def DOG(k=12, s=3):
    n = 2 * k + 1
    s1, s2 = s * np.sqrt(2), s / np.sqrt(2)
    kernel = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            kernel[i, j] = np.exp(-((i - k)**2 + (j - k)**2) / (2. * s1**2)) / (2 * np.pi * s1**2) - \
                           np.exp(-((i - k)**2 + (j - k)**2) / (2. * s2**2)) / (2 * np.pi * s2**2)
    kernel = np.round(kernel / np.sqrt((kernel**2).sum()), 3)
    return kernel

# Đường dẫn ảnh
img_paths = ["/kaggle/input/xla-images/elephant_g.jpg",
             "/kaggle/input/xla-images/aero.jpg",
             "/kaggle/input/xla-images/earthfromsky.jpg"]

# Vòng lặp qua từng ảnh
for path in img_paths:
    # Mở ảnh và chuyển đổi thành ảnh grayscale
    img = Image.open(path).convert('L')
    img_np = np.array(img) / 255.0  # Chuẩn hóa

    # Áp dụng kernel LoG
    log_kernel = LOG()
    log_filtered = convolve2d(img_np, log_kernel, mode='same')

    # Áp dụng kernel DoG
    dog_kernel = DOG()
    dog_filtered = convolve2d(img_np, dog_kernel, mode='same')

    # Hiển thị kết quả
    pylab.figure(figsize=(20, 20))

    pylab.subplot(221)
    plot_kernel(log_kernel, 'LoG Kernel')

    pylab.subplot(222)
    pylab.title('Output Image with LoG', size=20)
    pylab.imshow(np.clip(log_filtered, 0, 1), cmap='gray')

    pylab.subplot(223)
    plot_kernel(dog_kernel, 'DoG Kernel')

    pylab.subplot(224)
    pylab.title('Output Image with DoG', size=20)
    pylab.imshow(np.clip(dog_filtered, 0, 1), cmap='gray')

    pylab.show()
```

Ảnh sau khi xử lý

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import pyramid_gaussian

# Tải ảnh
img = Image.open("/kaggle/input/xla-images/elephant_g.jpg")
image = np.array(img)  # Chuyển sang mảng NumPy để tương thích với skimage

# Lấy kích thước của ảnh gốc
orig_nrows, orig_ncols = image.shape[:2]

# Tạo Gaussian pyramid với hệ số downscale là 2, giữ nguyên các kênh RGB
pyramid = tuple(pyramid_gaussian(image, downscale=2, channel_axis=-1))

# Hiển thị từng cấp độ của pyramid
plt.figure(figsize=(20, 5))
for i, p in enumerate(pyramid, 1):
    plt.subplot(1, len(pyramid), i)
    plt.imshow(np.clip(p, 0, 1))  # Giới hạn giá trị để tránh vấn đề hiển thị
    plt.title(f"{p.shape[0]}x{p.shape[1]}")
    plt.axis('off')

plt.suptitle('Gaussian Pyramid', size=30)
plt.show()

# Tạo ảnh composite
compos_image = np.zeros((orig_nrows, orig_ncols + orig_ncols // 2, 3), dtype=np.double)
compos_image[:orig_nrows, :orig_ncols, :] = pyramid[0]

# Khởi tạo vị trí hàng cho mỗi cấp độ của pyramid
i_row = 0
for p in pyramid[1:]:
    nrows, ncols = p.shape[:2]
    # Kiểm tra xem có đủ không gian không, sau đó cập nhật i_row dựa trên nrows
    if i_row + nrows <= compos_image.shape[0]:
        compos_image[i_row:i_row + nrows, orig_ncols:orig_ncols + ncols] = np.clip(p, 0, 1)
    i_row += nrows

# Hiển thị ảnh composite
fig, axes = plt.subplots(figsize=(20, 20))
axes.imshow(compos_image)
axes.axis('off')
plt.show()
```
