```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.fft import fftshift, fft2

# Mở các bức ảnh từ đường dẫn đã chỉ định và chuyển sang ảnh xám (grayscale)
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Tạo danh sách chứa các bức ảnh
list_img = [Image.open(path).convert('L') for path in img_paths]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Vòng lặp để tính và hiển thị phổ của bộ lọc Gaussian cho từng ảnh
for i in range(3):
    # Chuyển ảnh thành mảng numpy
    img_array = np.array(list_img[i])
    
    # Kích thước ảnh
    h, w = img_array.shape
    
    # Tạo một Gaussian LPF trong không gian tần số
    # Xác định các chỉ số trục x và y
    x = np.linspace(-0.5, 0.5, w)
    y = np.linspace(-0.5, 0.5, h)
    X, Y = np.meshgrid(x, y)
    
    # Đặt sigma cho bộ lọc Gaussian
    sigma = 0.1
    gaussian_kernel = np.exp(-0.5 * (X**2 + Y**2) / sigma**2)
    
    # Tính phổ của ảnh bằng cách áp dụng DFT và dịch chuyển tần số thấp về trung tâm
    dft_image = fftshift(fft2(img_array))
    
    # Áp dụng Gaussian LPF lên phổ ảnh
    dft_image_filtered = dft_image * gaussian_kernel
    
    # Lấy phổ ngược của ảnh đã lọc và chuyển về miền không gian
    img_filtered = np.abs(np.fft.ifft2(fftshift(dft_image_filtered)))

    # Tạo một subplot cho mỗi ảnh
    plt.subplot(3, 3, i + 1)
    plt.imshow(np.log(1 + np.abs(dft_image)), cmap='gray')
    plt.title(f'Original Spectrum {i+1}', size=20)
    plt.axis('off')

    # Hiển thị phổ của bộ lọc Gaussian
    plt.subplot(3, 3, i + 4)
    plt.imshow(gaussian_kernel, cmap='gray')
    plt.title(f'Gaussian LPF Spectrum {i+1}', size=20)
    plt.axis('off')

    # Hiển thị ảnh sau khi áp dụng Gaussian LPF
    plt.subplot(3, 3, i + 7)
    plt.imshow(img_filtered, cmap='gray')
    plt.title(f'Filtered Image {i+1}', size=20)
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.1, hspace=0.3)

# Hiển thị tất cả các bức ảnh trong cửa sổ
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.fft import fftshift, fft2, ifft2

# Mở các bức ảnh từ đường dẫn đã chỉ định và chuyển sang ảnh xám (grayscale)
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Tạo danh sách chứa các bức ảnh
list_img = [Image.open(path).convert('L') for path in img_paths]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Vòng lặp để tính và hiển thị phổ của bộ lọc High-Pass cho từng ảnh
for i in range(3):
    # Chuyển ảnh thành mảng numpy
    img_array = np.array(list_img[i])
    
    # Kích thước ảnh
    h, w = img_array.shape
    
    # Tạo một High-Pass Filter trong không gian tần số
    # Xác định các chỉ số trục x và y
    x = np.linspace(-0.5, 0.5, w)
    y = np.linspace(-0.5, 0.5, h)
    X, Y = np.meshgrid(x, y)
    
    # Đặt sigma cho bộ lọc High-Pass (ngược lại với Gaussian LPF)
    sigma = 0.1
    gaussian_kernel = np.exp(-0.5 * (X**2 + Y**2) / sigma**2)
    hpf_kernel = 1 - gaussian_kernel  # High-pass kernel

    # Tính phổ của ảnh bằng cách áp dụng DFT và dịch chuyển tần số thấp về trung tâm
    dft_image = fftshift(fft2(img_array))
    
    # Áp dụng High-Pass Filter lên phổ ảnh
    dft_image_filtered = dft_image * hpf_kernel
    
    # Lấy phổ ngược của ảnh đã lọc và chuyển về miền không gian
    img_filtered = np.abs(ifft2(fftshift(dft_image_filtered)))

    # Tạo một subplot cho mỗi ảnh
    plt.subplot(3, 3, i + 1)
    plt.imshow(np.log(1 + np.abs(dft_image)), cmap='gray')
    plt.title(f'Original Spectrum {i+1}', size=20)
    plt.axis('off')

    # Hiển thị phổ của bộ lọc High-Pass
    plt.subplot(3, 3, i + 4)
    plt.imshow(hpf_kernel, cmap='gray')
    plt.title(f'High-Pass Filter Spectrum {i+1}', size=20)
    plt.axis('off')

    # Hiển thị ảnh sau khi áp dụng High-Pass Filter
    plt.subplot(3, 3, i + 7)
    plt.imshow(img_filtered, cmap='gray')
    plt.title(f'Filtered Image {i+1}', size=20)
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.1, hspace=0.3)

# Hiển thị tất cả các bức ảnh trong cửa sổ
plt.show()
```

```python
from PIL import Image
import numpy as np
import numpy.fft as fp
from scipy import ndimage
import matplotlib.pyplot as plt

# Đường dẫn đến các bức ảnh
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Vòng lặp qua từng ảnh
for i, path in enumerate(img_paths):
    # Mở ảnh và chuyển đổi thành ảnh grayscale
    img = Image.open(path).convert('L')  # Chuyển thành ảnh xám
    img_np = np.array(img)
    
    # Chuyển ảnh vào không gian tần số với FFT
    freq = fp.fft2(img_np)
    
    # Áp dụng bộ lọc Gaussian trên phổ tần số
    freq_gaussian = ndimage.fourier_gaussian(freq, sigma=4)
    
    # Chuyển ảnh về không gian thời gian (ảnh gốc) sau khi lọc
    img_lpf = fp.ifft2(freq_gaussian).real  # Lấy phần thực của ảnh

    # Tạo subplot và hiển thị ảnh đã qua bộ lọc
    plt.subplot(1, 3, i + 1)
    plt.imshow(img_lpf, cmap='gray')
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# Hiển thị tất cả các ảnh đã qua xử lý
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

# Đường dẫn đến các bức ảnh
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Tạo một cửa sổ hiển thị lớn
plt.figure(figsize=(18, 12))

# Vòng lặp qua từng ảnh
for i, path in enumerate(img_paths):
    # Mở ảnh và chuyển đổi thành ảnh grayscale
    img = Image.open(path).convert('L')
    img_np = np.array(img) / 255  # Chuẩn hóa về khoảng [0, 1]

    # Bước 1: Hiển thị ảnh gốc
    plt.subplot(3, 4, i * 4 + 1)
    plt.imshow(img_np, cmap='gray')
    plt.axis('off')
    plt.title(f'Original Image {i + 1}')

    # Bước 2: Chuyển đổi Fourier và hiển thị phổ tần số của ảnh gốc
    F1 = fftpack.fft2(img_np)
    F2 = fftpack.fftshift(F1)
    spectrum = (20 * np.log10(0.1 + np.abs(F2))).astype(int)

    plt.subplot(3, 4, i * 4 + 2)
    plt.imshow(spectrum, cmap='gray')
    plt.axis('off')
    plt.title(f'Spectrum {i + 1} (Original)')

    # Bước 3: Thêm nhiễu tuần hoàn vào ảnh
    noisy_img = img_np.copy()
    for n in range(noisy_img.shape[1]):
        noisy_img[:, n] += 0.2 * np.cos(0.1 * np.pi * n)

    plt.subplot(3, 4, i * 4 + 3)
    plt.imshow(noisy_img, cmap='gray')
    plt.axis('off')
    plt.title(f'Image {i + 1} with Sinusoidal Noise')

    # Bước 4: Hiển thị phổ tần số của ảnh nhiễu
    F1_noisy = fftpack.fft2(noisy_img)
    F2_noisy = fftpack.fftshift(F1_noisy)
    spectrum_noisy = (20 * np.log10(0.1 + np.abs(F2_noisy))).astype(int)

    plt.subplot(3, 4, i * 4 + 4)
    plt.imshow(spectrum_noisy, cmap='gray')
    plt.axis('off')
    plt.title(f'Spectrum {i + 1} (Noisy)')

# Hiển thị toàn bộ các hình ảnh
plt.tight_layout()
plt.show()
```

Ảnh sau khi xử lí

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.io import imread
import scipy.fft as fp
from scipy.stats import norm

# Đọc ảnh và chuyển sang grayscale, sau đó chuẩn hóa độ sáng
img_path = "/kaggle/input/xla-images/elephant_g.jpg"
    
im = 255 * rgb2gray(imread(img_path))

# Tạo kernel Gaussian thủ công
size_x, size_y = im.shape
x = np.linspace(-size_x / 2., size_x / 2., size_x)
y = np.linspace(-size_y / 2., size_y / 2., size_y)
gauss_kernel = np.outer(norm.pdf(x, scale=3), norm.pdf(y, scale=3))

# Chuyển ảnh và kernel sang miền tần số
freq = fp.fft2(im)
freq_kernel = fp.fft2(fp.ifftshift(gauss_kernel))  # Đây là H

# Áp dụng phép chập trong miền tần số để làm mờ ảnh
convolved = freq * freq_kernel  # Theo định lý tích chập
im_blur = fp.ifft2(convolved).real

# Chuẩn hóa ảnh làm mờ để đảm bảo giá trị pixel trong khoảng [0, 255]
im_blur = 255 * im_blur / np.max(im_blur)

# Khôi phục ảnh từ ảnh làm mờ bằng bộ lọc nghịch đảo
epsilon = 10**-6  # Hằng số tránh chia cho 0
freq = fp.fft2(im_blur)
freq_kernel = 1 / (epsilon + freq_kernel)  # Tránh chia cho 0
convolved = freq * freq_kernel
im_restored = fp.ifft2(convolved).real

# Chuẩn hóa ảnh khôi phục
im_restored = 255 * im_restored / np.max(im_restored)

# Hiển thị kết quả
plt.figure(figsize=(12, 12))
plt.gray()

# Hiển thị ảnh gốc
plt.subplot(221)
plt.imshow(im)
plt.title('Ảnh gốc')
plt.axis('off')

# Hiển thị ảnh làm mờ
plt.subplot(222)
plt.imshow(im_blur)
plt.title('Ảnh bị làm mờ')
plt.axis('off')

# Hiển thị ảnh đã khôi phục
plt.subplot(223)
plt.imshow(im_restored)
plt.title('Ảnh khôi phục với bộ lọc nghịch đảo')
plt.axis('off')

# Hiển thị sự khác biệt giữa ảnh gốc và ảnh đã khôi phục
plt.subplot(224)
plt.imshow(im_restored - im)
plt.title('Sai lệch giữa ảnh khôi phục và ảnh gốc')
plt.axis('off')

plt.show()
```

# **Image deconvolution với Wiener filter**

```python
from PIL import Image
import matplotlib.pyplot as plt

# Đường dẫn đến các bức ảnh
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(50, 30))

# Vòng lặp qua từng ảnh
for i, path in enumerate(img_paths):
    # Mở ảnh và chuyển đổi thành ảnh grayscale
    img = Image.open(path).convert('L')  # Chuyển thành ảnh xám
    
    # Tạo một subplot cho mỗi bức ảnh
    plt.subplot(1, 3, i + 1)
    
    # Hiển thị bức ảnh xám
    plt.imshow(img, cmap='gray')  # Đặt `cmap='gray'` để hiển thị ảnh xám
    
    # Tắt hiển thị trục tọa độ x và y để ảnh nhìn đẹp hơn
    plt.axis('off')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# Hiển thị tất cả các bức ảnh trong cửa sổ
plt.show()
```

Ảnh sau khi xử lí

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, restoration
from skimage.io import imread
from scipy.signal import convolve2d as conv2

# Đường dẫn đến các bức ảnh
img_paths = [
    "/kaggle/input/xla-images/elephant_g.jpg",
    "/kaggle/input/xla-images/aero.jpg",
    "/kaggle/input/xla-images/earthfromsky.jpg"
]

# Tạo một cửa sổ hiển thị với kích thước lớn
plt.figure(figsize=(20, 8))

# Vòng lặp qua từng ảnh
for i, path in enumerate(img_paths):
    # Đọc và chuyển đổi ảnh sang ảnh grayscale
    img = color.rgb2gray(imread(path))
    
    # Thiết lập bộ lọc điểm mờ (Point Spread Function - PSF)
    n = 7  # Kích thước của PSF
    psf = np.ones((n, n)) / n**2  # Kernel trung bình

    # Áp dụng làm mờ với PSF và thêm nhiễu Gaussian
    blurred_img = conv2(img, psf, 'same')
    noisy_blurred_img = blurred_img + 0.1 * img.std() * np.random.standard_normal(img.shape)

    # Khử nhiễu và làm mờ bằng bộ lọc Wiener không giám sát
    restored_img, _ = restoration.unsupervised_wiener(noisy_blurred_img, psf)

    # Tạo các subplot để hiển thị ảnh gốc, ảnh bị nhiễu và ảnh khôi phục
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 4), sharex=True, sharey=True)
    plt.gray()  # Chuyển sang chế độ ảnh xám cho tất cả các ảnh

    # Hiển thị ảnh gốc
    axes[0].imshow(img)
    axes[0].axis('off')
    axes[0].set_title('Ảnh gốc', size=15)

    # Hiển thị ảnh bị nhiễu và làm mờ
    axes[1].imshow(noisy_blurred_img)
    axes[1].axis('off')
    axes[1].set_title('Ảnh bị nhiễu và làm mờ', size=15)

    # Hiển thị ảnh đã khôi phục
    axes[2].imshow(restored_img)
    axes[2].axis('off')
    axes[2].set_title('Ảnh khôi phục tự điều chỉnh', size=15)

    fig.tight_layout()
    plt.show()
```
